```python
import os
import numpy as np
from matplotlib import pyplot as plt
import librosa
import librosa.display
%matplotlib inline
```

```python
from turtle import title


y, sr = librosa.load('./[0094] S1-PD1-01-SA.wav',
                     mono=True)
D = librosa.stft(y)  # STFT of y
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True)
                        
img1 = librosa.display.waveplot(y, sr=sr, ax=ax[0])

img2 = librosa.display.specshow(S_db, x_axis='s', y_axis='log', ax=ax[1])
ax[1].set(xlabel='Tempo (s)')
ax[1].set(ylabel='Frequência (Hz)')
ax[0].set(ylabel='Frequência (Hz)')

fig.colorbar(img1, ax=[ax[0]], format="%+2.f dB",location=None)
fig.axes[2].set_visible(False)
# To eliminate redundant axis labels, we'll use "label_outer" on all subplots:
for ax_i in ax:
    ax_i.label_outer()
fig.colorbar(img2, ax=[ax[1]], format="%+2.f dB")
fig.savefig('2-pd.png', dpi=400)
```
